```python
# We want to retain some of the state data so that we can use it for visualizations later
import pandas as pd
```

```python
nasa_df = pd.read_csv("../data/Cleaned/Master_NASA_1974_2024.csv")
emissions_df = pd.read_csv("../data/Cleaned/Master_Emissions.csv")
noaa_df = pd.read_csv("../Data/Cleaned/Master_NOAA_1975_2024.csv")


if 'YEAR' in noaa_df.columns:
    noaa_df.rename(columns={'YEAR': 'Year'}, inplace=True)

# Retain state-level data from NOAA
total_state_level_df = noaa_df[['Year', 'STATE', 'TOTAL_DAMAGE', 'TOTAL_DEATHS', 'TOTAL_INJURIES']]

# Aggregate NOAA data by state and year
state_level_agg = total_state_level_df.groupby(['Year', 'STATE'], as_index=False).sum()

# Merge with emissions and temperature datasets using 'Year'
state_level_dataset = (
    state_level_agg.merge(emissions_df, on='Year', how='left')
                  .merge(nasa_df, on='Year', how='left')
)
```

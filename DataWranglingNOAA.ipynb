```python
# Step 1: Remove columns where all values are NaN or 0 (excluding DAMAGE_CROPS)
columns_to_drop = [
    col for col in Exploredf.columns 
    if Exploredf[col].isna().all() or (Exploredf[col] == 0).all()
]
columns_to_drop = [col for col in columns_to_drop if col != "DAMAGE_CROPS"]  # Keep DAMAGE_CROPS
Exploredf_cleaned = Exploredf.drop(columns=columns_to_drop)

# Step 2: Remove any column related to "TOR" (tornado-related data)
Exploredf_cleaned = Exploredf_cleaned.drop(columns=[col for col in Exploredf_cleaned.columns if "TOR" in col])

# Step 3: Drop unnecessary columns
columns_to_remove = ['END_LAT', 'END_LON', 'DATA_SOURCE']
Exploredf_cleaned = Exploredf_cleaned.drop(columns=[col for col in columns_to_remove if col in Exploredf_cleaned.columns])

# Display the cleaned DataFrame
Exploredf_cleaned.head()
```

```python
# Function to convert damage values into numeric format
def convert_damage(value):
    if isinstance(value, str):
        value = value.upper().strip()
        if value.endswith('K'):
            return float(value.replace('K', '')) * 1_000
        elif value.endswith('M'):
            return float(value.replace('M', '')) * 1_000_000
        elif value.endswith('B'):
            return float(value.replace('B', '')) * 1_000_000_000
        else:
            try:
                return float(value)  # Convert if it's a pure number
            except ValueError:
                return None
    return None

# Convert damage columns
Exploredf_cleaned['DAMAGE_PROPERTY'] = Exploredf_cleaned['DAMAGE_PROPERTY'].apply(convert_damage)
Exploredf_cleaned['DAMAGE_CROPS'] = Exploredf_cleaned['DAMAGE_CROPS'].apply(convert_damage) if 'DAMAGE_CROPS' in Exploredf_cleaned.columns else 0

# Fill NaN values with 0 for consistency
Exploredf_cleaned[['DAMAGE_PROPERTY', 'DAMAGE_CROPS']] = Exploredf_cleaned[['DAMAGE_PROPERTY', 'DAMAGE_CROPS']].fillna(0)

# Step 4: Create aggregated columns
Exploredf_cleaned['TOTAL_DAMAGE'] = Exploredf_cleaned['DAMAGE_PROPERTY'] + Exploredf_cleaned['DAMAGE_CROPS']
Exploredf_cleaned['TOTAL_DEATHS'] = Exploredf_cleaned['DEATHS_DIRECT'] + Exploredf_cleaned.get('DEATHS_INDIRECT', 0)
Exploredf_cleaned['TOTAL_INJURIES'] = Exploredf_cleaned['INJURIES_DIRECT'] + Exploredf_cleaned.get('INJURIES_INDIRECT', 0)

# Display updated DataFrame
Exploredf_cleaned.head()
```

```python
# Step 5: Apply Natural Disaster Filters
Exploredf_disasters = Exploredf_cleaned[
    (Exploredf_cleaned['TOTAL_DEATHS'] >= 10) |  # Condition 1: 10+ total deaths
    (Exploredf_cleaned['TOTAL_INJURIES'] + Exploredf_cleaned['TOTAL_DEATHS'] >= 100) |  # Condition 2: 100+ affected
    (Exploredf_cleaned['TOTAL_DAMAGE'] >= 12_500_000)  # Condition 3: Estimated real damage ≥ $25M
]

# Display the filtered dataset
Exploredf_disasters.head()
```

```python
# Great now that we know about what one year of natural disasters looks like, we can aggregate all csvs from 1975 to 2024
```

```python
import glob

# Define the path where all CSV files are stored
csv_files_path = "../data/NOAA/*.csv"  # Adjust path based on your file location
all_csv_files = glob.glob(csv_files_path)  # Get all CSV file paths

# Function to convert damage values into numeric format
def convert_damage(value):
    if isinstance(value, str):
        value = value.upper().strip()
        if value in ['', ' ', 'UNKNOWN']:  # Handle empty or unknown values
            return 0
        try:
            if value.endswith('K'):
                return float(value.replace('K', '')) * 1_000
            elif value.endswith('M'):
                return float(value.replace('M', '')) * 1_000_000
            elif value.endswith('B'):
                return float(value.replace('B', '')) * 1_000_000_000
            else:
                return float(value)  # Convert if it's a pure number
        except ValueError:
            return 0  # Default to 0 if conversion fails
    return 0  # If value isn't a string, assume it's missing and return 0

# Function to process each CSV file
def process_noaa_csv(file_path):
    # Load dataset
    df = pd.read_csv(file_path, low_memory=False)
    
    # Step 1: Remove columns where all values are NaN or 0 (excluding DAMAGE_CROPS)
    columns_to_drop = [col for col in df.columns if df[col].isna().all() or (df[col] == 0).all()]
    columns_to_drop = [col for col in columns_to_drop if col != "DAMAGE_CROPS"]  # Keep DAMAGE_CROPS
    df_cleaned = df.drop(columns=columns_to_drop)

    # Step 2: Remove any column related to "TOR" (tornado-related data)
    df_cleaned = df_cleaned.drop(columns=[col for col in df_cleaned.columns if "TOR" in col])

    # Step 3: Drop unnecessary columns
    columns_to_remove = ['END_LAT', 'END_LON', 'DATA_SOURCE']
    df_cleaned = df_cleaned.drop(columns=[col for col in columns_to_remove if col in df_cleaned.columns])

    # Convert damage columns
    df_cleaned['DAMAGE_PROPERTY'] = df_cleaned['DAMAGE_PROPERTY'].apply(convert_damage)
    df_cleaned['DAMAGE_CROPS'] = df_cleaned['DAMAGE_CROPS'].apply(convert_damage) if 'DAMAGE_CROPS' in df_cleaned.columns else 0

    # Fill NaN values with 0 for consistency
    df_cleaned[['DAMAGE_PROPERTY', 'DAMAGE_CROPS']] = df_cleaned[['DAMAGE_PROPERTY', 'DAMAGE_CROPS']].fillna(0)

    # Step 4: Create aggregated columns
    df_cleaned['TOTAL_DAMAGE'] = df_cleaned['DAMAGE_PROPERTY'] + df_cleaned['DAMAGE_CROPS']
    df_cleaned['TOTAL_DEATHS'] = df_cleaned['DEATHS_DIRECT'] + df_cleaned.get('DEATHS_INDIRECT', 0)
    df_cleaned['TOTAL_INJURIES'] = df_cleaned['INJURIES_DIRECT'] + df_cleaned.get('INJURIES_INDIRECT', 0)

    # Extract year from dataset (if missing, extract from filename)
    df_cleaned['YEAR'] = df_cleaned['YEAR'] if 'YEAR' in df_cleaned.columns else int(file_path[-12:-8])  

    # Step 5: Apply Natural Disaster Filters
    df_disasters = df_cleaned[
        (df_cleaned['TOTAL_DEATHS'] >= 10) |  # Condition 1: 10+ total deaths
        (df_cleaned['TOTAL_INJURIES'] + df_cleaned['TOTAL_DEATHS'] >= 100) |  # Condition 2: 100+ affected
        (df_cleaned['TOTAL_DAMAGE'] >= 12_500_000)  # Condition 3: Estimated real damage ≥ $25M
    ]

    return df_disasters

# Process all files and merge them into one master dataset
master_noaa = pd.concat([process_noaa_csv(file) for file in all_csv_files], ignore_index=True)

# Save to CSV without setting YEAR as index (keeping it as a column for grouping)
master_noaa.to_csv("../data/NOAA/Master_NOAA_1975_2024.csv", index=False)

```
